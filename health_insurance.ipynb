 What are the most common policy categories and types among different age groups or accommodation types?

```python


# Assuming 'Upper_Age' is the column you want to use for categorizing into age groups
age_bins = [0, 18, 30, 40, 50, 60, 70, 80]  # Define your age bins
age_labels = ['<18', '19-30', '31-40', '41-50', '51-60', '61-70', '>70']  # Labels for the age bins
health_insurance_df['Age_Group'] = pd.cut(health_insurance_df['Upper_Age'], bins=age_bins, labels=age_labels, right=False)
```

```python


# Custom aggregation function
def single_mode(series):
    modes = series.mode()
    if len(modes) > 0:
        return modes[0]  # Return the first mode
    else:
        return None  # Return None if the series is empty or has no mode

# Group by 'Age_Group' and find the most common 'Reco_Policy_Cat' within each group using the custom function
most_common_policy_cat_by_age = health_insurance_df.groupby('Age_Group')['Reco_Policy_Cat'].agg(single_mode)

# Do the same for 'Reco_Insurance_Type'
most_common_policy_type_by_age = health_insurance_df.groupby('Age_Group')['RECO_INSURANCE_TYPE'].agg(single_mode)

print(most_common_policy_cat_by_age)
print(most_common_policy_type_by_age)
```

```python
plt.Figure(figsize=(10,6));
sns.countplot(x='Age_Group',hue='Reco_Policy_Cat',data=health_insurance_df,palette='Set2');
plt.title('Distribution of Recommended Policy Category by Age Group');
plt.xticks(rotation=45);
```

How does the recommended policy premium vary by age, city, and accommodation type?

```python


# Assuming health_insurance_df is your DataFrame
# Optionally, create age bins if continuous
age_bins = [18, 30, 45, 60, 75, 90]  # Example age bins
age_labels = ['18-29', '30-44', '45-59', '60-74', '75+']
health_insurance_df['Age_Group'] = pd.cut(health_insurance_df['Upper_Age'], bins=age_bins, labels=age_labels, include_lowest=True)
```
